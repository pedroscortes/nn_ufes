# Operando com Imagens Digitais

## Representação de Imagens em Preto-e-Branco e Tons de Cinza

Como ilustrado nas Figuras 1 e 2, imagens são representadas no computador como matrizes de **pixels**. Imagens de alta resolução parecem contínuas devido à grande quantidade de pixels, mas provavelmente você já observou que ao dar bastante zoom em uma imagem, as células individuais (o pixels) se tornam visíveis.

Inclusive, é comum dizer que imagens cujos pixels são visíveis estão *pixeladas*.

Em imagens em preto e branco (*black and white*) ou em tons de cinza (*grayscale*), cada pixel contém um número indicando quanto da cor branco existe na célula. Na maioria dos casos,  os valores dos pixels são armazeandos usando ou o tipo `unsigned int de 8 bits (uint8)` ou o tipo `float (float32 ou float64)`. No primeiro caso, os valores dos pixels variam de 0 (preto) a 255 (branco), enquanto no segundo, os valores variam de 0.0 (preto) a 1.0 (branco).

Observe nas figuras que tons claros tendem a 255 e tons escuros tendem a zero.

## Imagens Coloridas

Em imagens coloridas, mais de um número é necessário por pixel para representar a cor da célula. Existem diversos **sistemas de cores**, mas o mais comum é o **RGB** no qual cada pixel armazena 3 números representando a quantidade de vermelho (**R**ed), verde (**G**reen) e azul (**B**lue). A cor preta é representada por $(0, 0, 0)$ e o branco por $(255, 255, 255)$ ou $(1.0, 1.0, 1.0)$, dependendo do tipo numérico utilizado. Um pixel totalmente verde é representado por $(0, 255, 0)$ e o roxo é representado por $(255, 0, 255)$, isto é, quantidades iguais de vermelho e azul (assumindo `uint8`).

Como ilustrado na Figura 3, outra forma de visualizar esta organização é imaginar que a imagem é composta por três matrizes (chamadas tipicamente de **canais**), uma contendo os valores de vermelho, outra os valores de verde e outra os valores de azul. A imagem completa é a concatenação dos três canais (não confundir o sinal de `+` com a soma dos canais).

Uma imagem com 128 pixels de altura, 256 pixels de largura e 3 canais, seria representada por uma matriz com 128 linhas, 256 colunas e profundidade 3, ou seja, com dimensões $(128, 256, 3)$. O valor da célula $(64, 128, 1)$ na matriz representa a quantidade de verde no pixel do meio da imagem.

A OpenCV não possui uma função pré-construída para download de imagens da internet. Vamos fazer isto usando a biblioteca requests. Para instalar a biblioteca localmente, execute a célula abaixo.

O código abaixo demonstra como realizar o download de uma imagem e salvá-la no disco.

````python
import cv2
from urllib.request import urlopen
import matplotlib.pyplot as plt

def download_image(url):
    resp = urlopen(url)
    image_bytes = np.asarray(bytearray(resp.read()), dtype=np.uint8)
    return cv2.imdecode(image_bytes, cv2.IMREAD_COLOR)

url = "https://t3.ftcdn.net/jpg/11/62/92/42/360_F_1162924268_X9jGUvG7O4irouRjRXhJfkkjYyQDv6lF.jpg"
path = "chameleon.jpg"

# invoca a funcao de download
img_bgr = download_image(url)

# converte a imagem para RGB
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)

# mostra as imagens
plt.subplot(1, 2, 1)
plt.imshow(img_bgr);
plt.title("Imagem BGR")
plt.subplot(1, 2, 2)
plt.title("Imagem RGB")
plt.imshow(img_rgb);
````
